```python
import cv2
from vehicle_detector import VehicleDetector

image = cv2.imread("/home/joshu/LPR-proj/data/LPRdatasets/vehicleImages/testData/UK/frame_00087_jpg.rf.d2c6825c3d1ae202c0d2505c071428c9.jpg")

detector = VehicleDetector()
vehicles = detector.detect(image)

print(f"Detected {len(vehicles)} vehicles")

for i, v in enumerate(vehicles):
    cv2.imshow(f"vehicle_{i}_{v['class']}", v["crop"])

cv2.waitKey(0)
cv2.destroyAllWindows()
```
